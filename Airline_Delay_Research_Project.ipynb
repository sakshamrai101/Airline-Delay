# Abstract


Airplane delays can be very frustrating. Even if a traveler did not miss the plane, having to wait around can make the rush to get there on-time feel like a waste. As a result, we wanted to know if there was a simple way to predict an airline's tendency for delays. One under-researched area in this field was relating the financials of a company to their airplane delays. Using revenue and expenses as a proxy for their quality (higher quality flights should have lower delays), we want to ask if there is a relationship between the two, using financials to predict an airline's average delay. Furthermore, we wanted to know whether expenses or revenue worked better as a proxy.

To prove a relationship, we ran a linear regression as well creating a different machine learning model in case there is a relationship, but not linear. We used the Bureau of Transportation Statistic's publishing of delay data, as well as an MIT Report on the financials of the major airline companies for these tests. The former provided the total amount of time delayed for an airline at an airport each month, which we summed up and standarized using the total amount of flights of that airline at that airport. The latter gave us each company's yearly revenue and expenses scaled by the average amount of seats moved per mile (Available Seat Mile). In our exploratory data analysis, we verified the assumptions of linear regression, showing that our variables are normally distributed and roughly linear with each other. We also checked for outliers in the financial metrics.

Running linear regression on both expenses vs delay and revenue vs delay, we found that in both cases there is a very weak relationship. Essentially every additional million dollars in expense or revenue, there would half a minute reduction in average delay per flight across a year. This effect is ever so slightly stronger when using expenses. Using a gradient boosted regression model, we hope for it to learn the relationship that we could not initially see. However, the model had weak predictive power, for both expenses and revenue, once again showing there is likely little relationship between financials and airplane delay.

As a result, we find that it is very likely that there is little to no relationship between company expenses and delay and between company revenue and delay, due to our small p-values. Because the coefficient values are so low, it is also difficult to make conclusions on whether expenses or revenue works better as a proxy for company operational investment, as there is likely no relationship to compare to begin with.



# Research Question

Does the operational investment of major US airline companies, as measured by their average expenses and revenue per available seat mile have a relationship with the frequency and duration of flight delays at the airports they service? Additionally, does expense per available seat mile or revenue per available seat mile have better predictive power?

## Background and Prior Work

When traveling, flight delays can be problematic for both travelers and the airline. For the traveler, they risk wasting time stuck in an airport or potentially missing out on time-sensitive events on arrival. For the airlines, delays can cascade into interfering with other flight landings, full day delays, and losing customers.<a name="cite_ref-3"></a>[<sup>1</sup>](#cite_note-3) As a result, understanding the nature of delays would be immensely valuable for planning by travelers and airlines.

There are many previous works looking for patterns in flight delays. For example, Mitsokapas, et. al. 's research paper attempts to create a statistical model for the probability of a flight being delayed by increasing amounts of time.<a name="cite_ref-3"></a>[<sup>1</sup>](#cite_note-3) While the overall model does not provide any new insights, the authors found that airlines tend to produce the most similar delay distributions within itself. YY’s article, on the other hand, presents a possible predictive model for flight delays. They find that flight departure time is one of the most predictive features for delays.<a name="cite_ref-4"></a>[<sup>2</sup>](#cite_note-4)

Because the variable of departure time is often inflexible for travelers, investigating why companies develop unique delay patterns may have more practical implications. However, the precise process and logistics of airline companies that lead to delays is unavailable to the public, meaning a different representation for airline logistics is needed. One interpretation of the delay of an airplane is that it is a measure of its quality. For example, if a company tries to make flights as cheap as possible (through paying workers less, lower quality equipment, etc.), then their flight could be considered lower quality and expected to have higher delays on average.

As a result, we want to use the financials of an airline company as an indicator for the quality of their flights and test if it has a relationship to their average delays on arrival.

# Hypothesis


In contrast to general intuition, we hypothesize that airlines with higher profit margins may experience more delays and cancellations. This hypothesis stems from the idea that while these airlines are financially successful, their focus might be more on maximizing profits and less on operational efficiency. For instance, in pursuit of higher revenues, these airlines might operate with tighter schedules or overbook flights, leading to increased delays. As such, we predict that companies with higher expenses will have lower average delay, while companies with higher revenue will have lower average delay.

Additionally, high profits do not automatically translate to operational investments for enhancing customer satisfaction; these airlines might allocate their profits towards other areas like shareholder dividends or expanding routes, rather than improving training or maintenance. Therefore, despite their financial success, these airlines could face more operational challenges, including delays and cancellations, due to their strategic choices.

As a result, we also predict that expenses will have better predictive power than revenue for airplane delays.

## Section 1: Normality of Flight Delays

Here, if we want to use regression to determine if expenses causes delay, we need to verify if both datasets are normally distributed. Moreover, because our financial data only contains about 250 samples, we must verify that the way we calculate delay for each financial sample is also normally distributed. Our dataset contains both the amount of flights that were delayed as well as the amount of delay time at a given airport on a specific month. To start, we can check the general distribution of the delay times.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(rc={'figure.figsize':(10,5)})
sns.set_style('white')
```

```python
delay_time_count = sns.histplot(x='arr_delay', data=airline_cleaned)
delay_time_count.set_title('Delay Time Counts')
plt.xlim(0, 20000)
plt.show()
```

As shown above, our delay time data is heavily skewed to the left. One possible explanation is that it should be typical for there to be little to no delay. After all, the expectation should be that our flight would not be delayed. This would be especially so as each sample is separated by airline, month, year, and the airport they service. To remedy this skew, we can try first summing the delays of each airline across all of the airports and months they serviced. We do this assuming that the lower end values will disappear as they are summed with higher end values.

```python
# First, groupby carrier and year to sum delays on all airports and months

delay = airline_cleaned[['airport','carrier','year','month','arr_delay']]

yearly_sums = delay.groupby(by=['carrier','year'])['arr_delay'].sum()
yearly_sums = pd.DataFrame(yearly_sums)

yearly_sum_plot = sns.histplot(data=yearly_sums,x='arr_delay')
yearly_sum_plot.set_title('Delay Time Across Year and Carrier')
```

Despite summing the delays, it seems a non-insignificant portion of delays still fall below 10000, while there are many months where airlines are delayed by nearly 10x as much. Looking into the original dataset, one can briefly notice that higher arr_delay value seem to have higher number of arrival flights. To test this, we can make a scatterplot comparison between the number of arrival flights and arr_delay for each.

```python
num_flights_and_delays = sns.scatterplot(data=total_airlines,x='arr_flights',y='arr_delay')
num_flights_and_delays.set_title('Amount of Delay vs Num of Flights at Airports')
plt.show()
```

While the spread in the scatterplot is incredibly large, there does seem to some increasing trend between the amount of delay a carrier has at an airport and the amount of flights coming in, which makes sense. We will now scale all of the delay values by dividing it by the number of flights coming in.

```python
delay_scaled = airline_cleaned[['airport','carrier','year','month','arr_delay','arr_flights']]
delay_scaled['scaled_delay'] = delay_scaled.loc[:,'arr_delay'] / delay_scaled.loc[:,'arr_flights']
delay_scaled.head()
```

```python
scaled_delay_count = sns.histplot(data=delay_scaled,x='scaled_delay')
scaled_delay_count.set_title('Delay Time scaled by Num of Flights')
plt.axis([0, 300, 0, 4000])
plt.show()
```

The total delay data has shifted from looking like an exponential decay to looking more normal but still radically skewed right. This was observed in the scatterplot by the large amount of points. Looking deeper, this result also makes sense because delay data by default is skewed right, as it cannot have a negative value. As a result, it is like looking at half of a normal distribution. To fix this, applying a log10 + 0.1 transformation could work now that the scaled data more closely resembles a bell curve.

```python
delay_scaled['log_delay'] = np.log10(delay_scaled['scaled_delay']+0.1)

log_delay_plot = sns.histplot(data=delay_scaled,x='log_delay', kde=True)
log_delay_plot.set_title('Log Scaled Delay Time')
plt.axis([0, 5, 0, 4000])
plt.show()
```

The log transformation has given us a normal distribution when applied on the total delay time of an airline at an airport for a given month and for a given year. However, the expenses data is given yearly for each of the different airlines. To ensure we can combine these two datasets, we need to sum up the delays across every airport and every month of year for a given airline and apply this same log transformation.

```python
# Similar transformation as the second graph, but using scaled delay time instead of raw

yearly_scaled_sums = delay_scaled.groupby(by=['carrier','year'])['scaled_delay'].sum()
yearly_scaled_sums = pd.DataFrame(yearly_scaled_sums)

scaled_delay_sums_plot = sns.histplot(data=yearly_scaled_sums,x='scaled_delay',kde=True,bins=40)
scaled_delay_sums_plot.set_title('Scaled Delay Time by Year and Carrier')
plt.show()
```

```python
yearly_scaled_sums['log_delay_sum'] = np.log10(yearly_scaled_sums['scaled_delay'])
log_delay_sum_plot = sns.histplot(data=yearly_scaled_sums,x='log_delay_sum')
log_delay_sum_plot.set_title('Log Scaled Delay Time by Year and Carrier')
plt.show()
```

In the two above plots, we actually see that when we sum up the scaled data across each airport and month for each airline and year, the data is weakly skewed right. When we apply a log transformation to the sum of the scaled, it becomes skewed left. 

This can be remedied below. By first summing up the total amount of flights and total amount of delay and then dividing these sums, the new distribution for each airline's yearly total scaled delay is approximately normal. The former approach is likely biased by the lower volume airports. The latter approach likely removes the airport's bias by summing first, giving us a better normal distributed.

Furthermore, this latter approach of summing first also seemingly removes the need for a log transformation, allowing us to make broader conclusions from our later analysis. **More importantly, by using this transformation, we can only conclude results for the average delay per flight across a year for each airline.** 

```python
a = delay_scaled.groupby(by=['carrier','year'])[['arr_delay','arr_flights']].sum()
a = pd.DataFrame(a)
a['total_scaled_delays'] = a['arr_delay'] / a['arr_flights']

total_scaled_delay_plot = sns.histplot(data=a,x='total_scaled_delays',kde=True,bins=40)
total_scaled_delay_plot.set_title('Summed then Scaled Delay by Year and Carrier')
plt.show()
```

### Scaled Delay Data Cleaning

```python
name_mapping = {
    'American Airlines Inc.' : 'American',
    'Continental Air Lines Inc.' : 'Continental',
    'Delta Air Lines Inc.' : 'Delta',
    'Northwest Airlines Inc.' : 'Northwest',
    'United Air Lines Inc.' : 'United',
    'US Airways Inc.' : 'US Airways',
    'Southwest Airlines Co.' : 'Southwest',
    'JetBlue Airways' : 'jetBlue',
    'AirTran Airways Corporation' : 'AirTran',
    'Frontier Airlines Inc.' : 'Frontier',
    'Alaska Airlines Inc.' : 'Alaska',
    'Hawaiian Airlines Inc.' : 'Hawaiian',
    'Spirit Air Lines' : 'Spirit',
    'Allegiant Air' : 'Allegiant',
    'Virgin America' : 'Virgin America',
    'America West Airlines Inc.' : 'America West'
} 

def convert_name(name):
    return name_mapping[name]
```

```python
# Recreating the summed first, then scaled delay time data

summed_df = total_airlines.groupby(by=['carrier_name','year'])[['arr_delay','arr_flights']].sum()
yearly_delay = summed_df['arr_delay'] / summed_df['arr_flights'].rename('scaled_delay')
yearly_delay = yearly_delay.reset_index()

# Converting carrier name to match the financial data

yearly_delay['carrier_name'] = yearly_delay['carrier_name'].apply(convert_name)
yearly_delay = yearly_delay[yearly_delay['year'] <= 2018]
display(yearly_delay.shape)
```

### First Linear Regression Model for Expense

```python
import patsy
import statsmodels.api as sm

outcome, predictors = patsy.dmatrices('scaled_delay ~ expense', expenses_delay_data)
model = sm.OLS(outcome, predictors)
result = model.fit()
print(result.summary())
```

Here, we see that every $1 million increase in expense only results in a decreased of about 0.5 minutes per flight. Moreover, the p-value is very low in this case (0.014), meaning it is very unlikely that this result came about by chance. The R^2 value of 0.034 also indicates a very weak relationship to this very weak effect to begin with.

To better visualize the poor predictive power of the linear model, we can also run a different linear regression package to be able to plot the residuals, or errors in our prediction.

### Second Linear Regression Model for Expense

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

expenses_delay_data = expenses_delay_data.dropna()

X_expenses = expenses_delay_data[['expense']]
y_expenses_delay = expenses_delay_data['scaled_delay']
model_expenses = LinearRegression()
model_expenses.fit(X_expenses, y_expenses_delay)
```

## Gradient Boosting Regression Model

One slight possibility could be that the relationship was not linear and that there is in fact a different better fitting relationship.

In this section we will examine the relationship further using the Gradient Boosting Regressor model. The Gradient Boosting Regressor in scikit-learn is an advanced ensemble machine learning model used for regression tasks, which combines multiple weak learning models to form a stronger predictive model. We decided to try Gradient Boosting because we thought linear regression's simplicity might limit its performance in identifying any potential underlying patterns in our data.

### Expense

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_expenses, y_expenses_delay, test_size=0.2, random_state=42)

params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "min_weight_fraction_leaf": 0.1,
    "alpha": 0.1,
}

model = GradientBoostingRegressor(**params)
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)

train_err = mean_squared_error(y_pred_train, y_train)
test_err = mean_squared_error(y_pred_test, y_test)
```

```python
print(train_err)
print(test_err)
```

The code above trains and evaluates our model to predict airline expenses based on delay data. We split the dataset into training and testing subsets, with 80% of the data used for training and 20% for testing. The model is trained using specified parameters and its performance is evaluated using mean squared error on both the training and test datasets. As shown above our test error is lower that our training error, which was a strange result.

```python
# Scatter plot for actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_train, model.predict(X_train), alpha=0.5, label='Train')
plt.scatter(y_test, y_pred_test, alpha=0.5, label='Test')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)  # Diagonal line
plt.show()
```

The plot suggests that the regression model has a reasonable level of predictive accuracy, as most data points cluster around the line of perfect prediction, but there is a noticeable trend of the model underpredicting at higher actual values. This might indicate a bias in the model where it doesn't capture the trend perfectly for higher value ranges, or it just tends to make safer/more conservative predictions for those values.

```python
# Calculating residuals
train_residuals = y_train - model.predict(X_train)
test_residuals = y_test - y_pred_test

# Residual plot
plt.figure(figsize=(10, 6))
plt.scatter(model.predict(X_train), train_residuals, alpha=0.5, label='Train')
plt.scatter(y_pred_test, test_residuals, alpha=0.5, label='Test')
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.hlines(y=0, xmin=y_pred_train.min(), xmax=y_pred_train.max(), colors='k', lw=2)
plt.legend()
plt.show()
```

The residual plot indicates that the model has a relatively even spread, and there is no obvious trend amongst residuals.

# Ethics & Privacy

During the initial stages of our project, our primary dataset in consideration is from the U.S. Bureau of Transportation which contains a comprehensive log of all domestic flights delays  in the US since 2018. It includes information about different types of delays including weather, security, system delay, etc. Possible ethical concerns related that could be:

1. Bias and Representation: The dataset, while extensive, may not fully capture all the reasons for flight delays, specifically for  smaller airlines or less-traveled routes. There's potential for over-representation of dominant airline concerns or under-representation of issues faced by minority or economically disadvantaged groups who might be affected differently by such delays.
2. Data Privacy and Sensitivity: While the dataset might not include individual passenger details, the granular break-up of delay causes could be sensitive. For instance, consistent security delays at a particular airport might raise unwarranted concerns among passengers, even if these were minor or non-threatening issues. Other potential security concerns might arise as we along with the project. 
3. Methodological Ethics: Underlying reasons for potential conclusions from preliminary data analysis might be ignored.  For instance, system delays could be a technological glitch, understaffing, or other factors. Misinterpreting or oversimplifying such causes could lead to skewed or non-directive analysis. 
4. Financial Data Interpretation: Adding financial revenue data raises ethical concerns about how this information is interpreted. Misinterpretation could lead to unfair assumptions about an airline's service quality or operational efficiency. High revenue doesn't always mean better service, as it can be influenced by factors like route popularity or pricing strategies. It's important to analyze this data carefully to avoid misleading conclusions about the relationship between an airline's financial performance and flight delays.

In addition to this, other potential datasets that we could potentially use/merge would ideally contain information about Flight Times, Price Fare, Delay Duration, Arrival and Departure airport and specific airport information. There are potential ethical concerns attached to this too.. Firstly, the process of merging disparate datasets might lead to inconsistencies or biases, possibly skewing our findings. Furthermore, drawing inferred links between Price Fare and flight delays could inadvertently cast misperceptions about certain airlines' value propositions. On the privacy front, while individual identities aren't explicit, there's a subtle risk of indirectly identifying frequent travelers  through overlapped dataset details regarding frequently picked airlines to travel, frequent class to travel and whether its a cheap or an expensive one, etc. Finally, merging varied data can sometimes obfuscate the distinction between correlation and causation, leading to conclusions that might be misleading. As we progress, it's crucial that these ethical considerations remain at the forefront of our analytical work.
